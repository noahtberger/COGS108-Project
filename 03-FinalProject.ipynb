### Top 10000 Songs on Spotify 1950-Now
This dataset includes a list of around 10000 songs that are popular on Spotify that released in the past 75 years. Additionally, it includes Spotify metadata such as popularity, danceability, energy, valence, loudness, speechiness, acousticness, and instrumentalness. All but the loudness and popularity are float values that range from 0 to 1, with 1 meaning that the song has a higher factor of the given variable. These values will help to determine if a song is more upbeat and joyful or slower and sadder. Additionally, Spotify gives each song a popularity score that helps their algorithm determine whether or not to recommend a song to the listener. This value ranges from 0, not a very popular song, to 100, a very popular song. This value will help us to determine the genres of music that have higher popularity by comparing them to the artist_genre values. Additionally, there is a tempo variable that is in beats per minute, which will help us to verify that some songs are edm, or similar genres of music. The main downside of this data set is that most of the data comes from values that Spotify gave each song, and the determination of each value is kept hidden by Spotify.

```python
#load in the dataset and get a basic look
df_spotify = pd.read_csv("data/00-raw/top_10000_1950-now.csv")
df_spotify.head()
```

After viewing the data, it appears that the headers are case sensitive and have white space. Therefore, while removing unnecessary columns, I am also going to make the column names more standardized.
```python
#Finding the number of nulls in each column
missing_counts = df_spotify_tidy.isnull().sum()
print(missing_counts)
print('\n')
percent_missing = (df_spotify_tidy.isnull().sum() / len(df_spotify_tidy)) * 100
print(percent_missing)
print('\n')
null_rows = df_spotify_tidy.isnull().any(axis=1).sum()
print("Number of rows with a null value: ", null_rows)
```

It appears that this dataset does not contain many null values, which means that it should be fine to remove all rows that contain a null in the cleaning process. We will still have 95% of our observations.

The only important outliers for this data are going to be in the lyrics column. This boxplot shows that there are mainly outliers with higher-than-average word counts. Most of the outliers can be written off as songs that simply have a lot of lyrics, like rap, or longer songs. However, there are some songs that have over 10000 words, which is most likely a data error, so we will remove them in the cleaning process. Additionally, there are some duplicates as well, so we will also remove them.

```python
#function to ensure that the lyrics are in english
def get_language(text):
    try:
        return detect(text)
    except:
        return 'idk'

#cleaning the data
#Removing duplicates
df_billboard_clean = df_billboard_tidy.drop_duplicates(subset=['year','ranking'], keep='first')
#ensures that the song isnt ridiculously long or too short
df_billboard_clean = df_billboard_clean[df_billboard_clean['word_count'] < 1200] #I determined that the songs with higher word count were a little messed up
df_billboard_clean = df_billboard_clean[df_billboard_clean['word_count'] > 15]
#Make sure everything is in english
df_billboard_clean['language'] = df_billboard_clean['lyrics'].apply(get_language)
df_billboard_clean = df_billboard_clean[df_billboard_clean['language'] == 'en']
rows = df_billboard_clean.shape[0]
columns = df_billboard_clean.shape[1]
print("The number of rows is: ", rows)
print("The number of columns is: ", columns)
print("The longest song is: ", df_billboard_clean['word_count'].max(), " words")
print("The shortest song is: ", df_billboard_clean['word_count'].min(), " words")
df_billboard_clean.head()
```

We showed earlier that we have no relevant nulls, and after removing duplicate songs, unrealistically wordy songs, songs with few words, and songs that are not in English, our dataset is now clean and tidy. 

Since our project primarily focuses on 'escapism' using the sentiment of song lyrics, we will primarily focus on data set #2. Now that we have all of our different measures of economic climate, we can begin to form a compiled "Economic Stress index" we do this as leaving the state of the economy up to any one single variable is irresponsible so we determine to take many individual measures and compose them together.

```python
##before we attempt to compile our datum, notice most data is compiled monthly beginning in the 50s, however gas is weekly begining in the 90s

df_GAS_clean.index= pd.to_datetime(df_GAS_clean.index)
df_GAS_M = df_GAS_clean.resample("ME").mean()

#We already know hte general shape of the gas graph, no need for a reprint!
#ax=df_GAS_M["GASREGW"].plot(figsize=(22,7))
#ax.tick_params(axis='x',labelsize=20)
#plt.title("Gas prices over time")
#plt.xlabel("Year")
#plt.ylabel("per gallon average")
```

```python
#now that Gas is resampled by month, we can create a month column / time key so that the dfs can be merged into one for our composite scale
df_infl["month"] = df_infl.index.to_period("M")
df_UNEMP_clean["month"] = df_UNEMP_clean.index.to_period("M")
df_sent_clean["month"] = df_sent_clean.index.to_period("M")
df_GAS_M["month"] = df_GAS_M.index.to_period("M")
```

```python
econ = df_infl.merge(df_UNEMP_clean,on = 'month', how = "inner")
econ = econ.merge(df_sent_clean,on = 'month',how = 'inner')
econ = econ.merge(df_GAS_M,on = 'month', how = 'inner')
```

Now that we have our econ dataframe merged up, we need to take the normalized Z-score of everything within it! It is imperative since each measure is under differing scales so if we simply merge them as is, some will have a greater influence on the score than others. This is because one dataframe can have larger raw values than another. Take sentiment vs gas prices for example; sentiment is on a scale of roughly 60 to 110 whereas gas prices stay under double digits.

```python
from scipy.stats import zscore

econ = econ.rename(columns={
    "CPIAUCSL": "inflation",
    "UNRATE": "unemployment",
    "UMCSENT": "sentiment",
    "GASREGW": "gas_price"
})

econ["z_inflation"] = zscore(econ["inflation"])
econ["z_unemployment"] = zscore(econ["unemployment"])
econ["z_sentiment"] = zscore(econ["sentiment"])
econ["z_gas_price"] = zscore(econ["gas_price"])
econ["economic_stress_index"] = (
    econ["z_inflation"]
    + econ["z_unemployment"]
    + econ["z_gas_price"]
    - econ["z_sentiment"]
)
econ["economic_stress_index"] = zscore(econ["economic_stress_index"])
econ["date"] = econ["month"].dt.to_timestamp()
ax = econ.plot(x="date", y="economic_stress_index", figsize=(22,7), legend=False)
ax.tick_params(axis='x', labelsize=20)
plt.title("Composite Economic Stress Index Over Time")
plt.xlabel("Year")
plt.ylabel("Standardized Stress")
```

And there we have our compiled economic stress idex! It begins in 1992 as that is the earliest point where all the data frames that it is composed of have data available. Generally, a higher stress value reflects a worse off state of the economy, as one would presume.

### Keyword-based Theme Analysis and Economic Correlations

To explore the relationship between popular music and the economic climate, we constructed three word lists which encapsulate escapist sentiment. Each song in the Billboard dataset was scored based on the frequency of theme-related keywords appearing in its lyrics.

```python
### We will be creating a word list of common party words, romance words, and escapist words
    # We need to figure out a better way to get these keywords as right now they were chosen at random from the internet
party_words = ['party', 'dance', 'club', 'drunk', 'night', 'celebrate', 'wild', 'smoke']
romance_words = ['love', 'kiss', 'heart', 'hold', 'forever', 'babe', 'mine', 'baby', 'darling', 'together']
escape_words = ['free', 'fly', 'high', 'lose', 'numb', 'forget', 'away', 'escape', 'dream']

# Count the number of times theme-related words appear in a song's lyrics
def keyword_score(lyrics, keywords):
    lyrics = str(lyrics).lower()
    return sum(lyrics.count(word) for word in keywords)

# Applying scoring to each song per theme
billboard_df['party_score'] = billboard_df['clean_lyrics'].apply(lambda x : keyword_score(x, party_words))
billboard_df['romance_score'] = billboard_df['clean_lyrics'].apply(lambda x : keyword_score(x, romance_words))
billboard_df['escape_score'] = billboard_df['clean_lyrics'].apply(lambda x : keyword_score(x, escape_words))
```

```python
# Aggregate keyword scores by year to observe shifts over time
yearly_df = billboard_df.groupby('year')[['party_score', 'romance_score', 'escape_score']].mean().reset_index()
yearly_df
```

The dataframe above shows how the prevalence of each theme has shifted over time through Billboard charting songs. We will now visualize these trends to better identify patterns over time.

```python
# Plot the average keyword scores for each theme over time
plt.figure(figsize=(12, 6))
plt.plot(yearly_df['year'], yearly_df['party_score'], label='Party')
plt.plot(yearly_df['year'], yearly_df['romance_score'], label='Romance')
plt.plot(yearly_df['year'], yearly_df['escape_score'], label='Escape')
plt.xlabel('Year')
plt.ylabel('Average Score')
plt.title('Song Theme Trends Over Time')
plt.legend()
plt.show()
```

The chart reveals that romance-themed language has been the dominant theme throughout the 2000-2023 period, though it shows a clear long-term decline. Party and escape themes remain relatively stable and low throughout the period, with a notable party spike around 2010-2011 coinciding with the post-financial crisis recovery.
